# Hidden Markov Model Decoding based on k-fold cross-validation

# 1. Import all the necessary packages and libraries

```python
from __future__ import print_function, division
from builtins import range
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
import json
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

## 2. Random Normalization

This function returns random matrices of shape (d1, d2) where rows sum up to 1. Useful for HMM initialization


```python
def random_normalized(d1, d2):
    x = np.random.random((d1, d2))
    return x / x.sum(axis=1, keepdims=True)
```

## 3. Design the Hidden Markov Model with k-fold cross validation

```python
class HMM:
    def __init__(self, M):
        self.M = M  # number of hidden states
        self.pi = None
        self.A = None
        self.B = None
        self.train_accs = []  # add empty list to track train accuracies
        self.test_accs = []  # add empty list to track test accuracies

    def fit(self, X, max_iter=10):
        np.random.seed(100)
        V = max(max(x) for x in X) + 1
        N = len(X)

        self.pi = np.ones(self.M) / self.M  # initial state distribution
        self.A = random_normalized(self.M, self.M)  # state transition matrix
        self.B = random_normalized(self.M, V)  # output distribution

        print("initial A:", self.A)
        print("initial B:", self.B)

        hidden_states_list = []
        costs = []
        for it in range(max_iter):
            if it % 10 == 0:
                print("it:", it)
            alphas = []
            betas = []
            scales =[]
            hidden_states = []
            logP = np.zeros(N)
            for n in range(N):
                x = X[n]
                T = len(x)
                scale = np.zeros(T)
                alpha = np.zeros((T, self.M))
                alpha[0] = self.pi * self.B[:, x[0]]
                scale[0] = alpha[0].sum()
                alpha[0] /= scale[0]
                for t in range(1, T):
                    alpha_t_prime = alpha [t - 1].dot(self.A) * self.B[:, x[t]]
                    scale[t] = alpha_t_prime.sum()
                    alpha[t] = alpha_t_prime /scale[t]
                logP[n] = np.log(scale).sum()
                alphas.append(alpha)
                scales.append(scale)

                beta = np.zeros((T, self.M))
                beta[-1] = 1
                for t in range(T - 2, -1, -1):
                    beta[t] = self.A.dot(self.B[:, x[t + 1]] * beta[t + 1])
                betas.append(beta)

                gamma = np.zeros((T, self.M))
                for t in range(T):
                    gamma[t] = alphas[n][t] * betas[n][t] / scales[n][t]
                hidden_states.append(np.argmax(gamma, axis=1))

            hidden_states_list.append([state.tolist() for state in hidden_states])

            with open('hidden_states_list.txt', 'w') as f:
                f.write(json.dumps(hidden_states_list))

            cost=np.sum(logP)
            costs.append(cost)

                # now re-estimate pi, A, B
            self.pi = sum((alphas[n][0] * betas[n][0]) for n in range(N)) / N
                # print "self.pi:", self.pi
                # break

            den1 = np.zeros((self.M, 1))
            den2 = np.zeros((self.M, 1))
            a_num = np.zeros((self.M, self.M))
            b_num = np.zeros((self.M, V))
            for n in range(N):
                x = X[n]
                T = len(x)
                den1 += (alphas[n][:-1] * betas[n][:-1]).sum(axis=0, keepdims=True).T
                den2 += (alphas[n] * betas[n]).sum(axis=0, keepdims=True).T

        # numerator for A
                for i in range(self.M):
                    for j in range(self.M):
                        for t in range(T - 1):
                            a_num[i, j] += alphas[n][t, i] * self.A[i, j] * self.B[j, x[t + 1]] * betas[n][t + 1, j]


            # numerator for B
                for i in range(self.M):
                    for t in range(T):
                        b_num[i, x[t]] += alphas[n][t, i] * betas[n][t, i]
            self.A = a_num / den1
            self.B = b_num / den2

        print("A:", self.A)
        print("B:", self.B)
        print("pi:", self.pi)

        plt.plot(range(len(costs)), np.array(costs))
        plt.xlabel('Iterations')
        plt.ylabel('Cost')
        plt.show()

    def log_likelihood(self, x):
        T = len(x)
        scale =np.zeros(T)
        alpha = np.zeros((T, self.M))
        alpha[0] = self.pi*self.B[:,x[0]]
        scale[0] = alpha[0].sum()
        alpha[0] /= scale[0]
        for t in range(1, T):
            alpha_t_prime = alpha[t - 1].dot(self.A.T) * self.B[:, x[t]]
            scale[t] = alpha_t_prime.sum()
            alpha[t] = alpha_t_prime / scale[t]
        return np.log(scale).sum()

    def log_likelihood_multi(self, X):
        return np.array([self.log_likelihood(x) for x in X])

    def get_state_sequence(self, x):
        T = len(x)
        delta = np.zeros((T, self.M))
        psi = np.zeros((T, self.M))
        delta[0] = self.pi * self.B[:, x[0]]
        for t in range(1, T):
            for j in range(self.M):
                delta[t, j] = np.max(delta[t - 1] * self.A[:, j]) * self.B[j, x[t]]
                psi[t, j] = np.argmax(delta[t - 1] * self.A[:, j])

# backtrack
        states = np.zeros(T, dtype=np.int32)
        states[T - 1] = np.argmax(delta[T - 1])
        for t in range(T - 2, -1, -1):
            states[t] = psi[t + 1, states[t + 1]]
        return states

    def accuracy(self, X, true_states):
        num_correct = 0
        total = 0
        for i in range(len(X)):
            predicted_states = self.get_state_sequence(X[i])[1]
            for j in range(len(predicted_states)):
                if predicted_states[j] == true_states[i][j]:
                    num_correct += 1
                total += 1
        return num_correct / total


def combined_HMM():
    X = []
    true_states = []
    for line in open('observed sequence.txt'):
        # 1 for H, 0 for T
        x = [1 if e == '1' else 2 if e == '2' else 3 if e == '3' else 4 if e == '4' else 5 if e == '5'else 6 if e == '6' else 7 if e =='7' else None for e
             in line.rstrip()]
        X.append(x)

    np.random.shuffle(X)

    split_index = int(len(X) * 0.8)
    X_train, X_test = X[:split_index], X[split_index:]

    hmm = HMM(3)
    hmm.fit(X_train)

    train_score = hmm.log_likelihood_multi(X_train).sum()
    print("Log-likelihood score for training set:", train_score)

    test_score = hmm.log_likelihood_multi(X_test).sum()
    print("Log-likelihood score for testing set:", test_score)

# k-fold cross validation 

    kf = KFold(n_splits=5, shuffle=True, random_state=10)
    fold = 5
    train_scores = []
    test_scores = []
    for train_index, test_index in kf.split(X):
        fold += 1
        print(f"Fold {fold}")

        X_train, X_test = [X[i] for i in train_index], [X[i] for i in test_index]

        hmm = HMM(3)
        hmm.fit(X_train)
        train_score = hmm.log_likelihood_multi(X_train).sum()
        test_score = hmm.log_likelihood_multi(X_test).sum()
        train_scores.append(train_score)
        test_scores.append(test_score)

    mean_train_score = np.mean(train_scores)
    std_train_score = np.std(train_scores)
    print(f"Mean train score: {mean_train_score}")
    print(f"Std train score: {std_train_score}")

    mean_test_score = np.mean(test_scores)
    std_test_score = np.std(test_scores)
    print(f"Mean test score: {mean_test_score}")
    print(f"Std test score: {std_test_score}")

# Plot the scores
    plt.figure(figsize=(10, 6))
    plt.title("Training and Testing Scores")
    plt.bar(["Training", "Testing"], [mean_train_score, mean_test_score], yerr=[std_train_score, std_test_score])
    plt.show()

    hmm.pi = np.array([0.4, 0.3, 0.4])
    hmm.A = np.array([[0.3, 0.3, 0.4], [0.4, 0.3, 0.3], [0.4, 0.2, 0.4]])
    hmm.B = np.array([[0.3, 0.4, 0.3], [0.5, 0.1, 0.4], [0.4, 0.3, 0.3]])

if __name__ == '__main__':
    combined_HMM()
```
